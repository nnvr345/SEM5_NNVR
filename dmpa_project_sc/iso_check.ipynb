```python
from sklearn.ensemble import IsolationForest
import pandas as pd
import numpy as np
import warnings; warnings.filterwarnings('ignore')
filepath='data_cleaning.csv'
df=pd.read_csv(filepath)
df.head()
```

```python
check_cols=['unit price',
    'line item quantity',
    'line item value',
    'weight (kilograms)',
    'freight cost (usd)',]
x_train=df[check_cols]
x_train.head()
```

```python
model=IsolationForest(n_estimators=100,max_samples='auto',contamination=float(0.2),max_features=1.0)
model.fit(x_train[['unit price']])
x_train['anomaly scores']=model.decision_function(x_train[['unit price']])
x_train['anomaly']=model.predict(x_train[['unit price']])
x_train
```
